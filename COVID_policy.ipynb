```
import pyspark.sql.functions as f
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# read in COVID lock-down data
lockdown = spark.read.option('header', True).csv('/FileStore/tables/U_S__State_and_Territorial_Stay_At_Home_Orders__March_15__2020___August_15__2021_by_County_by_Day.csv')
#lockdown.display()
```

```
#lockdown.printSchema()
```

```
lockdown = lockdown.withColumn("datetime", f.to_date(f.col("date"),"M/d/yyyy"))
#lockdown.display()
```

```
#lockdown.filter(f.col("Order_code") == "1").groupBy("State_Tribe_Territory", "Order_code", "Stay_at_Home_Order_Recommendation").agg(f.min("datetime")).display()
```

```
#lockdown.select(f.countDistinct("State_Tribe_Territory")).show()
```

```
lockdown2 = lockdown.groupBy("State_Tribe_Territory", "Order_code", "Stay_at_Home_Order_Recommendation").agg(f.min("datetime"))
#lockdown2.display()
```

```
small_order = lockdown.groupBy("State_Tribe_Territory").agg(f.min("Order_code")).withColumnRenamed("min(Order_code)", "Order_code")
#small_order.display()
```

```
test = small_order.join(lockdown2, on=["State_Tribe_Territory", "Order_code"]).withColumnRenamed("min(datetime)", "earliest_policy_date").orderBy("State_Tribe_Territory")

# remove AS
test = test.filter(f.col("State_Tribe_Territory")!="AS")
test.display()
```
